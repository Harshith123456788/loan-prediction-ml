**Problem Statement**

The goal of this project is to predict whether a loan will be approved or not based on applicant details such as income, education, credit history, loan amount, and other factors.

The dataset contains missing (null) values, so the first step is to clean the dataset by handling missing values. After preprocessing the data, different Machine Learning algorithms will be applied to predict the target variable (Loan_Status).

The model that provides the highest accuracy will be selected as the best-performing model.

Convsersion using LableEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']

for col in cols:
  df[col] = le.fit_transform(df[col])
```

```python
df.dtypes
```

** Define Feature Variables (X)**

X represents the input features used to train the model.
All columns except the target variable are included.
df.drop('Loan_Status', axis=1) removes the target column from the dataset.

** Define Target Variable (y)**

y represents the output variable the model needs to predict.
In this dataset, the target column is Loan_Status.

```python
X = df.drop(['Loan_Status','Loan_ID'], axis=1)
Y = df['Loan_Status']
```

```python
print(X)
```

```python
print(Y)
```

**Split the dataset into training and testing sets using train_test_split.**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

**Train-Test Split**

The dataset is divided into training data and testing data.

**Training Data (X_train, y_train)**

Used to train the machine learning model.

**Testing Data (X_test, y_test)**

Used to evaluate the model performance.

**test_size = 0.2**

20% of the data is used for testing.

80% of the data is used for training.

**random_state = 2**

Ensures the same data split every time the code runs.

Helps make the results reproducible.

**Logistic Regrssion **

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

**Accuracy Score**

It compares the actual loan status (y_test) with the predicted loan status (y_pred).

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, y_pred)

print(cm)
```

```python
import numpy as np

input_data = (1,1,0,1,0,5000,0,150,360,1,2)

# convert to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape for prediction
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)

print(prediction)

if prediction[0] == 1:
    print("Loan Approved")
else:
    print("Loan Not Approved")
```

**KNN algorithm**

The parameter n_neighbors determines the number of nearest neighbors used for classification.

The predicted class is determined by the majority vote of the nearest neighbors.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)

```

```python
y_pred_knn = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("KNN Accuracy:", accuracy_score(Y_test, y_pred_knn))
```

**Logistic Regression performed better than KNN for this dataset.**

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(Y_test, y_pred_knn))
```
